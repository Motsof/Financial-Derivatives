# Portafolio libre de riesgo

```python
w_sin_riesgo = [0.20,0.10,0.05,0.30,0.35]
r_sin_riesgo = np.array([0.20,0.10,0.05,0.30,0.35])
rend_port_rf = w_sin_riesgo @ r_sin_riesgo

print(rend_port_rf)
```

# Portafolio de tangencia

```python
print('Portafolio de tangencia o de mercado: \n')

vector_unos=[1,1,1,1,1]

matrix_cov_inv=np.linalg.inv(matrix_cov)

F=[i-rend_port_rf for i in media]

w_port_tan=(matrix_cov_inv @ F)/(vector_unos @ matrix_cov_inv @ F)

montos_port_tan=[capital*i for i in w_port_tan]

#numero de activos
n_activos=[math.floor(i/j) for i,j in zip(montos_port_tan,precios)]

#ultimo precio oservado
precios=[df_precios.loc['2024-04-24 00:00:00-06:00', df_precios.columns[i]]for i in range (len (df_precios.columns))]

#montos que tenemos que inveritr en cada activo
montos=[capital*i for i in w_port_tan]

#numero de activos
n_activos=[math.floor(i/j) for i,j in zip(montos,precios)]

#monto real
monto_real=[i*j for i,j in zip(precios,n_activos)]

port_tan=pd.DataFrame(zip(activos,precios,n_activos,monto_real,w_port_max_sharpe), columns=['activos','precios','# activos','monto','wi'])
print(port_tan)

#rendimiento de portafolio
rend_port_tan= media @ w_port_tan
print('\n Rendimiento del portafolio es de: ', rend_port_tan)

#volatilidad del portafolio
de_port_tan=np.sqrt(w_port_tan @ matrix_cov @ w_port_tan)
print('Volatilidad del portafolio es de: ', de_port_tan)

# cociente de sharpe
Sharpe_port_tan =rend_port_tan/de_port_tan
print('el cociente de sharpe es de:', Sharpe_port_tan)

#cociente de sharpe ajustado
sharpe_ajus_port_tan = (rend_port_tan - rend_port_rf)/de_port_tan
print(f"El portafolio tiene un cociente sharpe ajustado de: {sharpe_ajus_port_tan:.5}")

```
